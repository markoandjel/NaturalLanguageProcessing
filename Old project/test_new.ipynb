```python
CHOICES_BLOCK_RE = re.compile(r"choices': array\(\[(.*?)\]\s*,\s*dtype=object\)", re.S)
CHOICE_RE = re.compile(r"'((?:\\'|[^'])*)'")  # <-- handles \' inside strings

def extract_choices(raw: str):
    m = CHOICES_BLOCK_RE.search(raw)
    if not m:
        return []
    block = m.group(1)

    # Extract and unescape
    choices = [s.replace("\\'", "'") for s in CHOICE_RE.findall(block)]
    return choices
```

```python
mul_val = pd.read_csv("mc1_targets.csv")
mul_val["mc1_choices"] = mul_val["mc1_targets"].apply(extract_choices)

# sanity check row 5 (0-based index 4)
print(mul_val.loc[4, "mc1_choices"][3])
```
